2:1 IDENTIFY THE MOST COMMON COMBINATIONS OF CUISINES IN THE DATASET.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#import data
dataset= pd.read_csv("dataset.csv")
```

Data Preprocessing

```python
#check for null values
pd.isnull(dataset).sum()
```

```python
#drop all null values
dataset.dropna(inplace=True)
```

```python
#check database
dataset.shape
```

```python
dataset.info()
```

```python
#check description of data
dataset[['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']].describe()
```

```python
#count the occurences of each cuisine combination
cuisine_counts= dataset['Cuisines'].value_counts().head(10)

#display the most common cuisine combination
print("The most Common Cuisine Combinations:")
print(cuisine_counts)

#plot the top 10 cuisine combinations
plt.figure(figsize=(12,6))
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index, hue=cuisine_counts,palette="viridis")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combinations")
plt.title("Top 10 Most common Cuisines Combinations")
plt.show()

```

2:2 DETERMINE IF CERTAIN CUISINE COMBINATIONS TEND TO HAVE HIGHER RATINGS.

```python
#group by 'cuisines' and calculate the average rating
cuisine_ratings= dataset.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

#display the top 10 cuisine combinations with highest ratings
print("Top 10 Highest Rated Cuisine Combinations:")
print(cuisine_ratings)

#plot the top 10 cuisine combinations by rating
plt.figure(figsize=(10,6))
sns.barplot(x=cuisine_ratings.values, y=cuisine_ratings.index,hue=cuisine_ratings.index,palette="coolwarm")
plt.xlabel("Average Aggregate Rating")
plt.ylabel("Cuisine Combinations")
plt.title("Top 10 Highest Rated Cuisine Combinations")
plt.xlim(0,5)  #rating range from 0-5
plt.show()
```
